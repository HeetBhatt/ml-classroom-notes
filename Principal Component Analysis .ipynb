```python
from sklearn.decomposition import PCA 
from sklearn import datasets
from sklearn.preprocessing import LabelEncoder

iris_data = datasets.load_iris(as_frame=True)
l = LabelEncoder()
y = l.fit_transform(iris_data['target'])
data = iris_data.data
print(data.shape)
print(data.info())
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scaled_df = std_scaler.fit_transform(data)
pca = PCA(n_components=2)
x = pca.fit_transform(scaled_df)
print(pca.explained_variance_ratio_)
```

```python
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.scatter(
    x[:,0],
    x[:,1],
    c=y,
    cmap='rainbow',
    alpha=0.7,
    edgecolors='b'
)
```

```python
print(sum(pca.explained_variance_ratio_))
```

```python
# analyzing the variance ratio with 1 to 4 attributes 
nums = np.arange(1,5)
var_ratio = []
for num in nums:
    pca = PCA(n_components=num)
    pca.fit(scaled_df)
    var_ratio.append(np.sum(pca.explained_variance_ratio_))
print(var_ratio)
```

```python
plt.figure(figsize=(4,2),dpi=150)
plt.grid()
plt.plot(nums,var_ratio,marker='o')
plt.xlabel('n_components')
plt.ylabel('Explained variance ratio')
plt.title('n_components vs. Explained Variance Ratio')
```

```python
dataset2= datasets.load_breast_cancer(as_frame=True)
df2 = dataset2.data
print(df2.shape)
pca2 = PCA(n_components=1)
pca2.fit_transform(df2)
print(sum(pca2.explained_variance_ratio_))
```
