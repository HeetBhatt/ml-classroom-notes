```python
from sklearn.datasets import load_digits
import numpy as np
import pandas as pd 
```

```python
data = load_digits()
print(data.data)
```

```python
print(data.target[100])
plt.matshow(data.images[100])
plt.show()
```

```python
#Decision Tree 
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(data.data,data.target,test_size=30)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(criterion="entropy")
dtc.fit(X_train,y_train)
print(dtc.score(X_test,y_test))
```

```python
print(dtc.get_params())
```

```python
print(dtc.get_depth())
```

```python
import sklearn.tree as tree 
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(25,30))
_ = tree.plot_tree(dtc,feature_names=data.feature_names,class_names=str(data.target_names))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(criterion="entropy",n_estimators=200)
rfc.fit(X_train,y_train)
print(rfc.score(X_test,y_test))
```
